Query 6.2

Calculate average annual percentage return and volatility of all 30 stocks over a theoretical one year period

```python
close_price_dataframe.fillna(0,inplace = True)
close_price_dataframe.set_index(df26['Date'],inplace = True)

annual_return = close_price_dataframe.pct_change().mean()*100*252
newdf = pd.DataFrame(annual_return)
newdf.columns = ['Returns']
newdf['Volatility'] = close_price_dataframe.pct_change().std() * np.sqrt(252)
newdf
```

Query 6.3

Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method

```python
data = np.asarray([np.asarray(newdf['Returns']),np.asarray(newdf['Volatility'])]).T

errors = []

for k in range(1, 30):
    k_means = KMeans(n_clusters=k)
    k_means.fit(data)
    errors.append(k_means.inertia_)
fig = plt.figure(figsize=(15, 5))
plt.plot(range(1, 30), errors)
plt.grid(True)
plt.title('Elbow curve')

```

```python
k = 5

centroids,_ = kmeans(data,k)
idx,_ = vq(data,centroids)
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()

print(newdf.idxmax())

newdf.drop({'TATACHEM','IRCON'},inplace=True)

data = np.asarray([np.asarray(newdf['Returns']),np.asarray(newdf['Volatility'])]).T


centroids,_ = kmeans(data,k)
idx,_ = vq(data,centroids)
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()
```
